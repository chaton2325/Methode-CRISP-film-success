```python
#Chargement du dataset 
import pandas
df3 = pandas.read_csv('phases de traitement dataset/movie_dataset_final3.csv')
print(df3.info())
print(df3.describe())
print("Dimensions du dataset final nettoyé : ", df3.shape)
print("Tête du dataset : \n",df.head(5))
```

```python
# Import des librairies
#Attributs fortement corrélés avec la variable cible 'revenue' (Succès commmercial)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sélectionner uniquement les colonnes numériques
df_numeric = df3.select_dtypes(include=['int64', 'float64'])

# Afficher les premières lignes pour vérifier
print(df_numeric.head())

# Calculer la matrice de corrélation
correlation_matrix = df_numeric.corr()

# Affichage de la heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Matrice de corrélation des attributs numériques")
plt.show()

# Identifier les colonnes fortement corrélées avec 'revenue'
threshold = 0.5  # seuil pour considérer forte corrélation
strong_corr = correlation_matrix['revenue'][abs(correlation_matrix['revenue']) >= threshold].index.tolist()

# Retirer 'revenue' lui-même de la liste
strong_corr = [col for col in strong_corr if col != 'revenue']

print("Attributs fortement corrélés avec 'revenue' :", strong_corr)

# Optionnel : créer un DataFrame avec uniquement ces colonnes + 'revenue'
df_strong_corr = df_numeric[strong_corr + ['revenue']]
print(df_strong_corr.head())
```

```python
# Import des librairies
#Attributs fortement corrélés avec la variable cible 'Vote average' (Succès critique)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# On garde uniquement les colonnes numériques
df_numeric = df3.select_dtypes(include=['int64', 'float64'])

# Calcul de la matrice de corrélation
correlation_matrix = df_numeric.corr()

# Affichage de la heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Matrice de corrélation des attributs numériques")
plt.show()

# Identifier les attributs fortement corrélés avec 'vote_average'
threshold = 0.5  # seuil pour considérer forte corrélation
strong_corr = correlation_matrix['vote_average'][abs(correlation_matrix['vote_average']) >= threshold].index.tolist()

# Retirer 'vote_average' lui-même de la liste
strong_corr = [col for col in strong_corr if col != 'vote_average']

print("Attributs fortement corrélés avec 'vote_average' :", strong_corr)

# Optionnel : créer un DataFrame avec uniquement ces colonnes + vote_average
df_strong_corr = df_numeric[strong_corr + ['vote_average']]
print(df_strong_corr.head())
```

```python
#Analyse : On remarque que pour le succès commercial (revenu), les attributs fortement corrélés sont le budget et le nombre de votes. Cela suggère que les films avec un budget plus élevé et un plus grand nombre de votes ont tendance à générer des revenus plus élevés.

#Analyse : Pour le succès critique , On a aucuns attributs fortements corrélés avec la variable cible 'vote_average' (succès critique). Cela indique que le succès critique d'un film n'est pas fortement influencé par les autres attributs numériques disponibles dans le dataset.

#Conclusuion : On va crééer deux datasets finaux : un pour le succès commercial (On va supprimer un des attributs budget ou vote_count (Ici on va supprimer vote count car on aura bdesoin du budget pour prédire le succès commercial)) et un autre pour le succès critique (en gardant tous les attributs disponibles car aucun n'est fortement corrélé).
```

```python
#Maintenant on va visualiser les relations entre les attributs fortement corrélés et la variable cible 'revenue' à l'aide de graphiques de dispersion (scatter plots) pour mieux comprendre ces relations.
import pandas as pd
df4 = pd.read_csv('phases de traitement dataset/movie_dataset_final3.csv')
import matplotlib.pyplot as plt
# Graphique de dispersion pour 'budget' vs 'revenue'
plt.figure(figsize=(8, 6))
plt.scatter(df4['budget'], df4['revenue'], alpha=0.5)
plt.title('Budget vs Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
# Graphique de dispersion pour 'vote_count' vs 'revenue'
plt.figure(figsize=(8, 6))
plt.scatter(df4['vote_count'], df4['revenue'], alpha=0.5
)
plt.title('Vote Count vs Revenue')
plt.xlabel('Vote Count')
plt.ylabel('Revenue')
plt.show()

# Graphique de dispersion pour 'Runtime' vs 'revenue'
plt.figure(figsize=(8, 6))
plt.scatter(df4['runtime'], df4['revenue'], alpha=0.5)
plt.title('Runtime vs Revenue')
plt.xlabel('Runtime')
plt.ylabel('Revenue')
plt.show()

# Graphique de dispersion pour 'vote_average' vs 'revenue'
plt.figure(figsize=(8, 6))
plt.scatter(df4['vote_average'], df4['revenue'], alpha=0.5)
plt.title('Vote Average vs Revenue')
plt.xlabel('Vote Average')
plt.ylabel('Revenue')
plt.show()
```

```python
#Analyses des graphiques :
#On observe une tendance positive entre le budget et le revenu, indiquant que les films avec un budget plus élevé ont tendance à générer des revenus plus élevés. Cependant, la relation n'est pas strictement linéaire, et il y a une grande variabilité dans les revenus pour un budget donné.
#De meme que pour le nombre de votes.

#On observe que entre 70 et 180 minutes de durée, les films ont tendance à générer des revenus plus élevés, tandis que les films très courts ou très longs ont des revenus plus faibles en moyenne. Cela suggère qu'il y a une durée optimale pour les films en termes de succès commercial.

#Pour le vote average , on observe une légère tendance positive entre la note moyenne des votes et le revenu, mais la relation est moins prononcée que pour le budget et le nombre de votes. Cela suggère que bien que les films mieux notés puissent générer des revenus plus élevés, d'autres facteurs jouent également un rôle important dans le succès commercial d'un film.
```
