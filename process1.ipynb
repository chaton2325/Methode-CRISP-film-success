```python
# Import des librairies
#Attributs fortement corrélés avec la variable cible 'revenue' (Succès commmercial)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sélectionner uniquement les colonnes numériques
df_numeric = df3.select_dtypes(include=['int64', 'float64'])

# Afficher les premières lignes pour vérifier
print(df_numeric.head())

# Calculer la matrice de corrélation
correlation_matrix = df_numeric.corr()

# Affichage de la heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Matrice de corrélation des attributs numériques")
plt.show()

# Identifier les colonnes fortement corrélées avec 'revenue'
threshold = 0.5  # seuil pour considérer forte corrélation
strong_corr = correlation_matrix['revenue'][abs(correlation_matrix['revenue']) >= threshold].index.tolist()

# Retirer 'revenue' lui-même de la liste
strong_corr = [col for col in strong_corr if col != 'revenue']

print("Attributs fortement corrélés avec 'revenue' :", strong_corr)

# Optionnel : créer un DataFrame avec uniquement ces colonnes + 'revenue'
df_strong_corr = df_numeric[strong_corr + ['revenue']]
print(df_strong_corr.head())
```

```python
# Import des librairies
#Attributs fortement corrélés avec la variable cible 'Vote average' (Succès critique)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# On garde uniquement les colonnes numériques
df_numeric = df3.select_dtypes(include=['int64', 'float64'])

# Calcul de la matrice de corrélation
correlation_matrix = df_numeric.corr()

# Affichage de la heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Matrice de corrélation des attributs numériques")
plt.show()

# Identifier les attributs fortement corrélés avec 'vote_average'
threshold = 0.5  # seuil pour considérer forte corrélation
strong_corr = correlation_matrix['vote_average'][abs(correlation_matrix['vote_average']) >= threshold].index.tolist()

# Retirer 'vote_average' lui-même de la liste
strong_corr = [col for col in strong_corr if col != 'vote_average']

print("Attributs fortement corrélés avec 'vote_average' :", strong_corr)

# Optionnel : créer un DataFrame avec uniquement ces colonnes + vote_average
df_strong_corr = df_numeric[strong_corr + ['vote_average']]
print(df_strong_corr.head())
```

```python
#Analyse : On remarque que pour le succès commercial (revenu), les attributs fortement corrélés sont le budget et le nombre de votes. Cela suggère que les films avec un budget plus élevé et un plus grand nombre de votes ont tendance à générer des revenus plus élevés.

#Analyse : Pour le succès critique , On a aucuns attributs fortements corrélés avec la variable cible 'vote_average' (succès critique). Cela indique que le succès critique d'un film n'est pas fortement influencé par les autres attributs numériques disponibles dans le dataset.

#Conclusuion : On va crééer deux datasets finaux : un pour le succès commercial (On va supprimer un des attributs budget ou vote_count (Ici on va supprimer vote count car on aura bdesoin du budget pour prédire le succès commercial)) et un autre pour le succès critique (en gardant tous les attributs disponibles car aucun n'est fortement corrélé).
```

```python
#Creation des nouveaux datasets pour le succes critique et commercial avec les genres encodés
import pandas as pd

# Supposons que df_encoded est déjà chargé et contient les colonnes nécessaires, y compris genres encodés

# --- Dataset Succès Critique ---
df_critique = df_encoded.copy()

# Créer la colonne succes_critique : 1 si vote_average >= 6, sinon 0
df_critique['succes_critique'] = (df_critique['vote_average'] >= 6).astype(int)

# Supprimer la colonne vote_average
df_critique = df_critique.drop(columns=['vote_average'])

# Sauvegarder le dataset succès critique
output_file_critique = 'phases de traitement dataset/movie_dataset_critical_success.csv'
df_critique.to_csv(output_file_critique, index=False)
print(f"Dataset succès critique sauvegardé sous : {output_file_critique}")

# --- Dataset Succès Commercial ---
df_commercial = df_encoded.copy()

# Créer la colonne succes_commercial : 1 si revenue >= 2/3 * budget, sinon 0
df_commercial['succes_commercial'] = (df_commercial['revenue'] >= (2/3) * df_commercial['budget']).astype(int)

# Supprimer les colonnes vote_count et revenue (car on veut prédire succès commercial)
df_commercial = df_commercial.drop(columns=['vote_count', 'revenue'])

# Sauvegarder le dataset succès commercial
output_file_commercial = 'phases de traitement dataset/movie_dataset_commercial_success.csv'
df_commercial.to_csv(output_file_commercial, index=False)
print(f"Dataset succès commercial sauvegardé sous : {output_file_commercial}")
```

```python
import pandas as pd

def downsample_majority(df, target_col, ratio=1.2, random_state=42):
    """
    Réduit la classe majoritaire à `ratio * taille classe minoritaire`.
    df : DataFrame
    target_col : nom de la colonne cible binaire
    ratio : facteur multiplicatif (>1) pour la taille finale de la classe majoritaire
    random_state : graine pour reproductibilité
    """
    # Classes minoritaire et majoritaire
    df_minority = df[df[target_col] == 0]
    df_majority = df[df[target_col] == 1]

    # Taille cible pour la classe majoritaire
    n_minority = len(df_minority)
    n_majority_new = int(n_minority * ratio)

    # Downsampling aléatoire de la classe majoritaire
    df_majority_downsampled = df_majority.sample(n=n_majority_new, random_state=random_state)

    # Reconstituer dataset équilibré partiellement
    df_balanced = pd.concat([df_minority, df_majority_downsampled]).sample(frac=1, random_state=random_state).reset_index(drop=True)


    return df_balanced

#pour succès critique
df_critique = pd.read_csv('phases de traitement dataset/movie_dataset_critical_success.csv')
df_critique_balanced = downsample_majority(df_critique, target_col='succes_critique', ratio=1.2)
print("Distribution après équilibrage partiel succès critique :")
print(df_critique_balanced['succes_critique'].value_counts())

#pour succès commercial
df_commercial = pd.read_csv('phases de traitement dataset/movie_dataset_commercial_success.csv')
df_commercial_balanced = downsample_majority(df_commercial, target_col='succes_commercial', ratio=1.4)
print("Distribution après équilibrage partiel succès commercial :")
print(df_commercial_balanced['succes_commercial'].value_counts())

#sauvegarde 
df_critique_balanced.to_csv('phases de traitement dataset/movie_dataset_critical_success_balanced.csv', index=False)
df_commercial_balanced.to_csv('phases de traitement dataset/movie_dataset_commercial_success_balanced.csv', index=False)
```
